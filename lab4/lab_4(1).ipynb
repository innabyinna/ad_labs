```python
import os
import pandas as pd

def txt_to_csv(txt_file, csv_file, delimiter=';'):
    if os.path.exists(csv_file):
        os.remove(csv_file)

    dtype_spec = {
        'Global_active_power': 'float',
        'Global_reactive_power': 'float',
        'Voltage': 'float',
        'Global_intensity': 'float',
        'Sub_metering_1': 'float',
        'Sub_metering_2': 'float',
        'Sub_metering_3': 'float',
    }

    df = pd.read_csv(txt_file, delimiter=delimiter, dtype=dtype_spec, na_values='?')
    df.to_csv(csv_file, index=False)
    df.info()

# Шляхи до файлів
txt_file_path = 'household_power_consumption.txt'
csv_file_path = 'household_power_consumption.csv'

# Виклик функції
txt_to_csv(txt_file_path, csv_file_path)
```

```python
def pd_df(csv_file_path):
    dtype = {
        'Global_active_power': float,
        'Global_reactive_power': float,
        'Voltage': float,
        'Global_intensity': float,
        'Sub_metering_1': float,
        'Sub_metering_2': float,
        'Sub_metering_3': float
    }

    df = pd.read_csv(csv_file_path, dtype=dtype, na_values='?')
    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
    df.dropna(inplace=True)
    return df
csv_file_path = 'household_power_consumption.csv'
df = pd_df(csv_file_path)    
display(df.head())
print(df.isnull().sum())
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
def within_current_range(df):
    filtered_df = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20)]
    
    result_df = filtered_df[(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3'])]
    
    return result_df

start_time = time.time()
result_dataframe = within_current_range(df)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

NumPy:

```python
def within_current_range_np(nparray):
    filtered_np = nparray[(nparray[:, 3] >= 19) & (nparray[:, 3] <= 20)]
    result_np = filtered_np[(filtered_np[:, 7] > filtered_np[:, 8])]
    
    return result_np

start_time = time.time()
result_npdataframe = within_current_range_np(npdata)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Numpy: {end_time}")
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії

```python
def random_house_avarage(df, x=500000):
    random_indices = df.sample(n=x, replace=False).index
    random_raw = df.loc[random_indices]
    average_value = random_raw[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    
    return average_value

start_time = time.time()
average_value = random_house_avarage(df)
print(average_value)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

NumPy:

```python
def random_house_avarage_np(nparray, x=500000):
    random_indices = np.random.choice(np.arange(nparray.shape[0]), size=x, replace=False)
    random_raw = nparray[random_indices]
    average_value = np.mean(random_raw[:, 6:9], axis=0)
    
    return average_value

start_time = time.time()
average_value = random_house_avarage_np(npdata)
print(average_value)
end_time = time.time() - start_time
print(f"Час виконання Numpy: {end_time}")
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

```python
def select_households(df):
    filtered_df = df[(df['Time'] > datetime.time(18, 0, 0)) & (df['Global_active_power'] > 6.0)]

    final_df = filtered_df[(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_1']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3'])]
   
    first_half = final_df.head(len(final_df) // 2)
    second_half = final_df.tail(len(final_df) // 2)
    result_df = pd.concat([first_half.iloc[::3, :], second_half.iloc[::4, :]])
    return result_df

start_time = time.time()
result = select_households(df)
display(result.head())
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

NumPy:

```python
def select_households_np(nparray):
    
    filtered_np = nparray[(nparray[:, 1]  > datetime.datetime.strptime('18:00:00', '%H:%M:%S').time()) & (nparray[:, 2] > 6.0)]
    final_np = filtered_np[(filtered_np[:, 7] > filtered_np[:, 6]) & (filtered_np[:, 7] > filtered_np[:, 8])]
   
    first_half = final_np[:len(final_np) // 2]
    second_half = final_np[len(final_np) // 2:]
    result_np = np.concatenate([first_half[::3], second_half[::4]])
    
    return result_np

start_time = time.time()
result = select_households_np(npdata)
print(result)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```
